```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
ride_df = pd.read_csv("ride_data.csv")
city_df = pd.read_csv("city_data.csv")

city_df = city_df.drop(city_df.index[84])

    




```

```python

city_df["total rides"] = np.array(ride_df.groupby("city").ride_id.count())
city_df['average fare'] = np.array(ride_df.groupby("city").fare.mean())
driver_count = city_df["driver_count"]
array_types = city_df["type"]
color_key = {"Urban" : "lightcoral","Rural" : "gold" ,"Suburban": "lightskyblue" }
array_types = array_types.map(color_key)
type_urban_df = city_df[city_df.type == "Urban"]
type_rural_df = city_df[city_df.type == "Rural"]
type_suburban_df = city_df[city_df.type == "Suburban"]
city_df.head()
```

```python
plt.scatter(type_urban_df["total rides"], type_urban_df["average fare"], 
            s = type_urban_df['driver_count'],color = "lightcoral", label = "Urban", edgecolor = "black", linewidths = .3)
plt.scatter(type_rural_df["total rides"], type_rural_df["average fare"], 
            s = type_rural_df['driver_count'],color = "gold", label = "Rural", edgecolor = "black", linewidths = .3)
plt.scatter(type_suburban_df["total rides"], type_suburban_df["average fare"], 
            s = type_suburban_df['driver_count'],color = "lightskyblue", label = "Suburban", edgecolor = "black", linewidths = .3)
plt.xlabel("Total Numbe of Rides(Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data(2016)")
plt.legend()
plt.show()
```

```python
num_types = []
num_types.append(city_df["city"][city_df.type == "Urban"].count())
num_types.append(city_df["city"][city_df.type == "Rural"].count())
num_types.append(city_df["city"][city_df.type == "Suburban"].count())
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
plt.pie(num_types, labels = labels, shadow= "true",
        autopct = "%1.1f%%", colors = colors, explode = (0,0.1,0))
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
num_rides = []
num_rides.append(city_df["total rides"][city_df.type == "Urban"].sum())
num_rides.append(city_df["total rides"][city_df.type == "Rural"].sum())
num_rides.append(city_df["total rides"][city_df.type == "Suburban"].sum())
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
plt.pie(num_rides, labels = labels, shadow= "true",
        autopct = "%1.1f%%", colors = colors, explode = (0,0.1,0))
plt.title("% of Total Rides by City Type")
plt.show()
```

```python
num_drivers = []
num_drivers.append(city_df["driver_count"][city_df.type == "Urban"].sum())
num_drivers.append(city_df["driver_count"][city_df.type == "Rural"].sum())
num_drivers.append(city_df["driver_count"][city_df.type == "Suburban"].sum())
labels = ["Urban", "Rural", "Suburban"]
colors = ["lightcoral", "gold", "lightskyblue"]
plt.pie(num_drivers, labels = labels, shadow= "true",
        autopct = "%1.1f%%", colors = colors, explode = (0,0.1,0))
plt.title("% of Total Drivers by City Type")
plt.show()
```
